**WHAT IS AIRBNB**

The company has been criticized for enabling bait-and-switch scams, being involved in West Bank settlements, possibly driving up home rents and creating nuisances for those living near leased properties. **What can we learn form predications ?(ex; location ,prices,reviews,etc)**

```python
areas_reviews = new_df.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
areas_reviews
```

```python
area=areas_reviews['neighbourhood_group']
review=areas_reviews['number_of_reviews']

fig=plt.figure(figsize=(10,5))
# bar plot created
plt.bar(area,review,color='maroon',width=0.4)
plt.xlabel("area")
plt.ylabel("review")
plt.title("area vs no of reviews")
plt.show()
```

```python
price_area=new_df.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head(5)
```

```python
area=price_area['price']
price=price_area['number_of_reviews']
fig=plt.figure(figsize=(10,5))
# create bar plot
plt.scatter(area,price)
plt.xlabel("price")
plt.ylabel("number of review")
plt.title("price vs number of reviews")
plt.show()
```

**A. Where the customer pays the highest and lowest rent according to location ?**

```python
airbnb_df = df.dropna()
```

```python
airbnb_price_df = airbnb_df.copy()
```

#Customer pays highest rent in location



```python
max_price_df=airbnb_price_df.groupby('neighbourhood_group',as_index=False)['price'].max().sort_values(['price'],ascending=False).rename(columns={'price':'Maximum price','neighbourhood_group':'location'})
max_price_df
```

# customer pays lowest rent

```python

min_price_df=airbnb_price_df.groupby('neighbourhood_group',as_index=False)['price'].min().sort_values(['price'],ascending=True).rename(columns={'price':'Minimum price','neighbourhood_group':'location'})
min_price_df
```

#Get the combine dataframe of minimum and maximum price according to location

```python
Merge_price_df=pd.merge(max_price_df,min_price_df,on='location')
Merge_price_df
```
